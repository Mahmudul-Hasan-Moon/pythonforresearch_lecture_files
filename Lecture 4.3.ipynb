### Lebel Encoding

```python
import category_encoders as ce

import pandas as pd
```

```python
train_df=pd.DataFrame({'Degree':['High school','Masters','Diploma','Bachelors','Bachelors','Masters','Phd','High school','High school']})
```

```python
train_df
```

```python
encoder = ce.OrdinalEncoder(cols=['Degree'], return_df = True, mapping = [{'col':'Degree', 'mapping':
                                                                          {'None':0, 'High school':1, 'Diploma':2, 'Bachelors':3, 'Masters':4, 'Phd':5}}])

train_df
```

```python
df = encoder.fit_transform(train_df)

df
```

### One hot encoding

```python
import category_encoders as ce

import pandas as pd
```

```python
data=pd.DataFrame({'City':[
'Delhi','Mumbai','Hydrabad','Chennai','Bangalore','Delhi','Hydrabad','Bangalore','Delhi'
]})
data
```

```python
encoder = ce.OneHotEncoder(cols = 'City', handle_unknown = 'return_nan', return_df=True, use_cat_names = True )

data
```

```python
df = encoder.fit_transform(data)

df
```
