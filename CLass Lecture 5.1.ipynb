```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [50,60,52,55,62]

plt.plot(x,y, color='red', linewidth=5, linestyle = 'dotted')

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Chart")
```

```python
plt.scatter(x,y)
```

```python
days = [1,2,3,4,5,6]
max_t = [50,51,52,53,54,48]
min_t = [42,41,43,40,38,33]
avg_t = [42,41,43,40,35,33]

plt.plot(days,max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="avg")

plt.legend(loc = "best")

plt.grid()
```

```python
import numpy as np

product = np.array([1,2,3,4])
price = [15,5,40,30]
profit = [5,3,10,5]

plt.bar(product,price,label = "Price", color="red", width=.4)
plt.bar(product, profit, label = "Profit", width=.4)

plt.legend()
```

```python
import numpy as np

product = np.array([1,2,3,4])
price = [15,5,40,30]
profit = [5,3,10,5]

plt.barh(product-.2,price,label = "Price", color="red")
plt.barh(product+.3, profit, label = "Profit")

plt.legend()
```
